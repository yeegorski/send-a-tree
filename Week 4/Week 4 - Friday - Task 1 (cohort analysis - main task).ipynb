The developers might have had a small bug in the first week that could have affected the user experience for those users who sent a free tree.
The task is to create a cohort analysis and check the retention numbers of those users who sent their first free_tree during this period (between Sep 13, 2019 and Sep 19, 2019) and compare it to the retention numbers of other cohorts of users who sent their first free_tree when the product was fixed (e.g. between Sep 20, 2019 and Sep 26, 2019).
This cohort analysis is aimed to investigate whether the users have returned in any of the first 7 days after their first free_tree send.
When the analysis is performed, it is to be visualized in Google Data Studio. 

```python
import csv
from datetime import date
from datetime import timedelta 
import pandas as pd
```

```python
user_sends = {}
with open('/home/dataguy/free_tree/first21_free.csv') as csvfile:
    readCSV = csv.reader(csvfile, delimiter=' ')
    for row in readCSV:
        if row[2] not in user_sends:
            user_sends[row[2]] = []
        if row[1] not in user_sends[row[2]]:    
            user_sends[row[2]].append(row[1])
```

```python
user_sends_list = list(user_sends.values())
```

```python
user_sends_list
```

```python
#a function to convert dates from the datetime format to the string 
def conv_date_str(dt):
    if dt.month<10: #some entries here to make sure we get a standardized output with zeroes in months and days
        return str(dt.year)+'-0'+str(dt.month)+'-'+str(dt.day)
    elif dt.day<10:
        return str(dt.year)+'-'+str(dt.month)+'-0'+str(dt.day)
    elif dt.month<10 and dt.day<10:
        return str(dt.year)+'-0'+str(dt.month)+'-0'+str(dt.day)
    else:
        return str(dt.year)+'-'+str(dt.month)+'-'+str(dt.day)
```

```python
#a function to convert dates from the string format to the datetime to make date calculations simpler 
def conv_date(dt):
    year = int(dt[:4])
    month = int(dt[5:7])
    day = int(dt[8:])
    return date(year,month,day)
```

```python
#a function that calculates the day differnce between two dates
def daydelta(dt1,dt2):
    x = conv_date(dt1)
    y = conv_date(dt2)
    delta = x - y
    return delta.days
```

```python
#creating a list of dates (start date - Sep 13, 2019)
date_list = []
x = date(2019,9,13)
for i in range(0,21):
    date_list.append(conv_date_str(x))
    x=x+timedelta(days=1)
```

```python
date_list
```

```python
count=0
cohort_sum = pd.DataFrame(columns=['cohort','days_passed','number_of_users']) #creating the dataframe
for cohort in date_list[0:14]: #cohorts of the first two weeks
    for i in range(0,8): #days passed since the original cohort date
        count = 0
        for j in range(0,len(user_sends_list)): #checking the number of comebacks
            if (date_list[i+date_list.index(cohort)] in user_sends_list[j]) and (user_sends_list[j][0]==cohort):
                count+=1
        row = {'cohort':cohort, 'days_passed':daydelta(date_list[i+date_list.index(cohort)],cohort), 'number_of_users':count}
        cohort_sum = cohort_sum.append(row, ignore_index=True)       
```
